```python
import random
import math
import matplotlib.pyplot as plt

def main():
  n = 1000
  values = []
  for i in range(n):
    x = random.random()
    y = random.random()
    values.append([x, y])
  inCircle = 0
  for value in values:
    if value[0] * value[0] + value[1] * value[1] < 1:
      inCircle = inCircle + 1
  calcPi = inCircle * 4 / n
  print("n = " + str(n) + "  pi = {pi:.6f}".format(pi = calcPi) + "  error = {err:.4f}".format(err = abs((math.pi - calcPi) / math.pi) * 100) + "%")
  n = 10000
  values = []
  for i in range(n):
    x = random.random()
    y = random.random()
    values.append([x, y])
  inCircle = 0
  for value in values:
    if value[0] * value[0] + value[1] * value[1] < 1:
      inCircle = inCircle + 1
  calcPi = inCircle * 4 / n
  print("n = " + str(n) + "  pi = {pi:.6f}".format(pi = calcPi) + "  error = {err:.4f}".format(err = abs((math.pi - calcPi) / math.pi) * 100) + "%")
  inCircleX = []
  inCircleY = []
  outOfCircleX = []
  outOfCircleY = []
  for point in values:
    if point[0] * point[0] + point[1] * point[1] < 1:
      inCircleX.append(point[0])
      inCircleY.append(point[1])
    else:
      outOfCircleX.append(point[0])
      outOfCircleY.append(point[1])
  plt.scatter(inCircleX, inCircleY, 1, color = "blue", marker = "o")
  plt.scatter(outOfCircleX, outOfCircleY, 1, color = "red", marker = "o")
  plt.show()
  n = 100000
  values = []
  for i in range(n):
    x = random.random()
    y = random.random()
    values.append([x, y])
  inCircle = 0
  for value in values:
    if value[0] * value[0] + value[1] * value[1] < 1:
      inCircle = inCircle + 1
  calcPi = inCircle * 4 / n
  print("n = " + str(n) + "  pi = {pi:.6f}".format(pi = calcPi) + "  error = {err:.4f}".format(err = abs((math.pi - calcPi) / math.pi) * 100) + "%")
  n = 1000000
  values = []
  for i in range(n):
    x = random.random()
    y = random.random()
    values.append([x, y])
  inCircle = 0
  for value in values:
    if value[0] * value[0] + value[1] * value[1] < 1:
      inCircle = inCircle + 1
  calcPi = inCircle * 4 / n
  print("n = " + str(n) + "  pi = {pi:.6f}".format(pi = calcPi) + "  error = {err:.4f}".format(err = abs((math.pi - calcPi) / math.pi) * 100) + "%")
  
main()
```
